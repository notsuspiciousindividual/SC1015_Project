# Create 2DF for ALL winners and another for further cleaning to select only some winners

```python
bestPic2 = winners[(winners['category'] == "BEST PICTURE") | (winners['category'] == "OUTSTANDING PICTURE") | (winners['category'] == "OUTSTANDING PRODUCTION") | (winners['category'] == "OUTSTANDING MOTION PICTURE") | (winners['category'] == "BEST MOTION PICTURE")]

someWinners = winners_some_clean[(winners_some_clean['category'] == "BEST PICTURE") | (winners_some_clean['category'] == "OUTSTANDING PICTURE") | (winners_some_clean['category'] == "OUTSTANDING PRODUCTION") | (winners_some_clean['category'] == "OUTSTANDING MOTION PICTURE") | (winners_some_clean['category'] == "BEST MOTION PICTURE") | (winners_some_clean['category'] == "ACTOR") | (winners_some_clean['category'] == "ACTRESS") | (winners_some_clean['category'] == "ACTOR IN A LEADING ROLE") | (winners_some_clean['category'] == "ACTRESS IN A LEADING ROLE")]
someWinners
```
